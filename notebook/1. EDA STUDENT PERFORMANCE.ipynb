```python
df['total_score'] = df['math_score'] + df['reading_score']+ df['writing_score']
df['average'] = df['total_score']/3
df.head()
```

```python
reading_full = df[df['reading_score']==100]['average'].count()
writing_full = df[df['writing_score']==100]['average'].count()
math_full = df[df['math_score']==100]['average'].count()

print(f'Number of student with full marks in Maths: {math_full}')
print(f'Number of student with full marks in Writing: {writing_full}')
print(f'Number of student with full marks in Reading: {math_full}')
```

```python
reading_less_20 = df[df['reading_score']<=20]['average'].count()
writing_less_20 = df[df['writing_score']<=20]['average'].count()
math_less_20 = df[df['math_score']<=20]['average'].count()

print(f'Number of student with less then 20 marks in Maths: {math_less_20}')
print(f'Number of student with less then 20 marks in Reading: {reading_less_20}')
print(f'Number of student with less then 20 marks in Writing: {writing_less_20}')
```

```python
gender_group = df.groupby('gender').mean()
gender_group
```

```python
plt.figure(figsize=(10,8))

X = ['Total Average', 'Math Average']

female_score = [gender_group['average'][0], gender_group['math_score'][0]]
male_score = [gender_group['average'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_score, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_score, 0.4, label = "Female")

plt.xticks(X_axis, X)
plt.ylabel('Marks')
plt.title("Total Average v/s Math average marks of both the genders", fontweight = 'bold')
plt.legend()
plt.show()
```
